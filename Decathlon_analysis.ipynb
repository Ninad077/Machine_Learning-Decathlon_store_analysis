## <font color = blue>1. Data Preprocessing</font>

I have created 2 excel files for the yearwise data of Decathlon segregating the excel file 'customer_transactions_sample.xlsx'

The  **first file** contains the yearwise data of Decathlon for the Year **'2009_2010'**.

The  **second file** contains the yearwise data of Decathlon for the Year **'2010_2011'**.

Let us import the libraries & read those datasets first.

### <font color = red>1.1 Importing the libraries</font>

```python
#Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### <font color = red>1.2 Reading the data</font>

```python
# Let us look at the '2009_2010_sample' data

df1= pd.read_excel('2009_2010_sample.xlsx')
```

```python
df1
```

```python
# Let us look at the '2010_2011_sample' data

df2 = pd.read_excel('2010_2011_sample.xlsx')
```

```python
df2
```

### <font color = red>1.3 Appending the data</font>

```python
# Appending the excel files of 'df1' & 'df2' using concat function.

data = pd.concat([df1, df2], ignore_index=True)
```

```python
data
```

### <font color = red>1.4 Checking for null values</font>

```python
# Now let us check for any null values in the dataset

data.isnull().sum()
```

```python
# We have got around '4382' null values in Description & '243007' null values in Customer ID. 
# Let us treat the null values for the columns 'Description' & 'Customer_ID' with it's mode.


data["Description"].fillna(data["Description"].mode()[0],inplace =True)
data["Customer ID"].fillna(data["Customer ID"].mode()[0],inplace =True)
```

```python
# Now let us check for null values again

data.isnull().sum()
```

### <font color = red>1.5 Formatting the data types of columns</font>

```python
# Convert InvoiceDate to datetime

data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
```

```python
# Ensure Quantity and Price are numeric

data['Quantity'] = pd.to_numeric(data['Quantity'])
data['Price'] = pd.to_numeric(data['Price'])
```

```python
# Let us check the cleaned data now.

data
```

## <font color = blue>3. Performing Time Series analysis (Holt_Winter model) to forecast the price for next 6 months.</font>

Let us do more analysis and forecast the price for the next 6 months on the given data using Reinforcement machine 
learning aka Time Series analysis.

### <font color = red>1. Previewing the dataset.</font>

```python
# Reading the dataset first 
 
data
```

### <font color = red>2. Filtering unnecessary columns.</font>

```python
# Now since we are only concerned about forecasting Price for InvoiceDate we will keep only those columns.

data = data[['InvoiceDate','Price']]
data
```

```python
# Checking the dimensions
data.shape
```

### <font color = red>3. Sorting InvoiceDate</font>

```python
# We sort the InvoiceDate column in ascending order and check for null values.
data = data.sort_values('InvoiceDate', ascending=True)
data
```

```python
data.isnull().sum()
```

```python
# Let us first check the value count for Order date
data['InvoiceDate'].value_counts
```

### <font color = red>4. Grouping InvoiceDate</font>

```python
# As one can clearly see we have multiple entries for the same 'Invoicedate'. To avoid using multiple entries we
# groupby function.

data = data.groupby('InvoiceDate')['Price'].sum().reset_index()
data
```

```python
# Thus the shape has changed, since we grouped the data. It has reduced from 1067371 rows × 9 columns to 
# 47635 rows × 2 columns
data.shape
```

### <font color = red>5. Setting InvoiceDate as Index</font>

```python
# Setting InvoiceDate as Index
# Since we want to forecast on just the Price column we are setting InvoiceDate column as an Index

data = data.set_index('InvoiceDate')
```

```python
data
```

### <font color = red>6. Resampling the data</font>

```python
# Resampling the data
# We resample the data to avoid inconsistencies & gapping. The Holt winter model won't work well if we have gapping
# within the data

# Let us keep the resampled data in 'y'
y = data['Price'].resample('M').mean()
```

```python
y
```

```python
print(y.shape)
```

```python
# Visualising 'y'

y.plot(figsize=(10,6))
plt.show()
```

### <font color = red>7. Splitting 'y' into train & test data</font>

```python
# Splitting the 'y' data into train & test data

# It is time to split the y data into Train & test. We have kept all the values before '2011-02-01' as Train data 
# & all the values after '2011-03-01' as Testing data.
# Loc is used here to locate the index position for the Date specified.

train = y.loc[ :'2011-10-01']
print(train.shape)
train
```

```python
test = y.loc['2011-10-30':]
print(test.shape)
test
```

```python
# Plotting the train & test data

train.plot(figsize=(10,6),title='Price over months',fontsize = 14, label='Train data')
test.plot(figsize=(10,6),title='Price over months',fontsize = 14, label='Test data')
plt.legend()
plt.show()
```

The above figure clearly shows the divison of train & test data. The Train data is represented with blue line & the Test
data is represented using orange line.

### <font color = red>7. Applying Holt_Winter model</font>

```python
# Applying Triple Exponential Smoothing (Holt Winter model)

# We introudce here 3 hyperparameters: smoothing line, smoothing slope & seasonal period. A seasonal period= 6 indicates 
# 6 months.

from statsmodels.tsa.api import ExponentialSmoothing
Holt_Winter_df = test.copy()

# smoothing level = beta
# aplha value should be in the range of 0 to 1
# values close to 0 indicate that older values are very less compared to the recent values.
# values close to 1 indicate that older values are equally important.

fit1 = ExponentialSmoothing(train, seasonal_periods=6, trend='add', seasonal='add').fit()

Holt_Winter_df['Holt_linear'] = fit1.forecast(steps=len(test))

train.plot(figsize=(10,6), title='Price', fontsize=14, label='Train data')
test.plot(figsize=(10,6), title='Price', fontsize=14, label='Test data')
Holt_Winter_df['Holt_linear'].plot(figsize=(10,6), title='Price', fontsize=14, label='Predicted Price')
plt.legend()
plt.show()
```

### <font color = red>8. Evaluating the model</font>

```python
# Checking RMSE for Holt Winter model.

from sklearn.metrics import mean_squared_error, r2_score
from math import sqrt

# Compute RMSE
rmse = sqrt(mean_squared_error(test, Holt_Winter_df['Holt_linear']))

# Compute R-squared (R^2)
r_squared = r2_score(test, Holt_Winter_df['Holt_linear'])

print(f'RMSE: {rmse}')
print(f'R-squared (R^2): {r_squared}')
```

Here the value of RMSE is quite less indicating the error in actual & predicted results is very less.
One can confirm from the plot of Holt_winter model above by comparing the predicted result price (Green) with actual
test data (orange) and looking  how closely predicted results imitate the test results. One can improve the accuracy even
more by alloting more training data while splitting. 

Forecasting will help Decathlon to have a rough estimate of the Price for forthcoming years. This can help the retail
shops to manage its inventory, time its import and regulate it's supply chain judicially.
